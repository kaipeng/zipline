```python
%pylab inline
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import HTML

import pytz
from datetime import datetime

from zipline.data.loader import load_bars_from_yahoo
from zipline.data.loader import load_bars_from_bloomberg

# Load data manually from Yahoo! finance
startdate = datetime(2001, 1, 1, 0, 0, 0, 0, pytz.utc)
enddate = datetime(2014, 11, 1, 0, 0, 0, 0, pytz.utc)
#secs=['S5TELS INDEX','S5ENRS INDEX']

secs=['S5TELS INDEX','S5ENRS INDEX','S5FINL INDEX','S5HLTH INDEX','S5UTIL INDEX','S5COND INDEX','S5INFT INDEX','S5INDU INDEX','S5CONS INDEX','S5MATR INDEX']
#data = load_bars_from_yahoo(stocks=secs, start=startdate, end=enddate)

fields=['BEST_PE_RATIO', 'BEST_CUR_EV_TO_EBITDA', 'BEST_PX_BPS_RATIO']
options={"OVERRIDE-BEST_FPERIOD_OVERRIDE":"BF"}
data = load_bars_from_bloomberg(securities=secs, fields=fields, options=options, start=startdate, end=enddate)
print data
```

```python
from zipline.algorithm import TradingAlgorithm
from zipline.api import order_target, record, symbol, history, add_history
import numpy as np

from collections import deque

algo_name = "Bloomberg Data + Fundamentals - EBITDA more extreme"

# Define algorithm
def initialize(context):
    #set_benchmark(symbol('SPY'))
                  
    # Register 2 histories that track daily prices,
    # one with a 100 window and one with a 300 day window
    #add_history(10, '1d', 'price')
    add_history(50, '1d', 'BEST_CUR_EV_TO_EBITDA')
    add_history(100, '1d', 'BEST_CUR_EV_TO_EBITDA')
    add_history(200, '1d', 'BEST_CUR_EV_TO_EBITDA')

    context.i = 0
    context.secs = secs
    
    #context.window = pd.Panel(items=context.secs)
    #for sec in context.secs:
    context.cache = {}
    #context.cache['xs_ma'] = deque()
    context.cache['s_ma'] = deque()
    context.cache['m_ma'] = deque()
    context.cache['l_ma'] = deque()
        

def handle_data(context, data):
    # Skip first 300 days to get full windows
    context.i += 1
    if context.i < 400:
        return
    
    # Progess report
    if context.i % 100 == 0:
        print 'Bars Completed: ', context.i, 'Data: ', data[data.keys()[0]]['dt'], 'Pos Value: ', context.portfolio.positions_value,'PnL: ', context.portfolio.pnl 

    # Compute averages
    # history() has to be called with the same params
    # from above and returns a pandas dataframe.
    s_hist = history(50, '1d', 'BEST_CUR_EV_TO_EBITDA')
    m_hist = history(100, '1d', 'BEST_CUR_EV_TO_EBITDA')
    l_hist = history(200, '1d', 'BEST_CUR_EV_TO_EBITDA')
    
    short_mavg = s_hist.mean()
    med_mavg = m_hist.mean()
    long_mavg = l_hist.mean()
    
    short_std = s_hist.std()
    med_std = m_hist.std()
    long_std = l_hist.std()
    
    #short_z = short_mavg/short_std
    #med_z = med_mavg/med_std
    #long_z = long_mavg/long_std
    
    short_z = {}
    med_z = {}
    long_z = {}
    
    threshold_lower = -2
    threshold_upper = 2
    threshold_lower2 = -3
    threshold_upper2 = 3
    for sec in data.keys():
        short_z[sec] = (data[sec].BEST_CUR_EV_TO_EBITDA - short_mavg[sec]) / short_std[sec]
        med_z[sec] = (data[sec].BEST_CUR_EV_TO_EBITDA - med_mavg[sec]) / med_std[sec]
        long_z[sec] = (data[sec].BEST_CUR_EV_TO_EBITDA - long_mavg[sec]) / long_std[sec]
        ##### Trading logic ######
        # Compounding Moving Average Crossover Signals
        target = 0
        if short_z[sec] > threshold_upper:
            target -= 20
        if med_z[sec] > threshold_upper:
            target -= 20
        if long_z[sec] > threshold_upper:
            target -= 20
        if short_z[sec] < threshold_lower:
            target += 30
        if med_z[sec] < threshold_lower:
            target += 40
        if long_z[sec] < threshold_lower:
            target += 50

        if short_z[sec] > threshold_upper2:
            target -= 20
        if med_z[sec] > threshold_upper2:
            target -= 20
        if long_z[sec] > threshold_upper2:
            target -= 20
        if short_z[sec] < threshold_lower2:
            target += 30
        if med_z[sec] < threshold_lower2:
            target += 40
        if long_z[sec] < threshold_lower2:
            target += 50            

        ##### Trading logic ######
        
        # order_target orders as many shares as needed to
        # achieve the desired number of shares.
        order_target(sec, target)

        # Save values for later inspection
        record(sec, data[sec].price)
        record(sec+'-factor', data[sec].BEST_CUR_EV_TO_EBITDA)
        record(sec+'-s_ma', short_mavg[sec])
        record(sec+'-m_ma', med_mavg[sec])
        record(sec+'-l_ma', long_mavg[sec])
    
    
    context.cache['s_ma'].append(short_mavg)
    context.cache['m_ma'].append(med_mavg)
    context.cache['l_ma'].append(long_mavg)
    
    #trim cache
    for key in context.cache.keys():
        if len(context.cache[key]) > 3:
            context.cache[key].popleft()
        

# Create algorithm object passing in initialize and
# handle_data functions
algo_obj = TradingAlgorithm(initialize=initialize, 
                            handle_data=handle_data)

# Run algorithm
perf_manual = algo_obj.run(data)
perf_manual.to_pickle(algo_name+'.pickle')
```

```python
import pandas as pd

def file_analyze(filename, begin=0, end=-1):
    data = load_bars_from_yahoo(stocks=['SPY'], start=startdate,
                            end=enddate)
    #data['SPY'].ix[:,:-2].plot(title='SPY', figsize=(15, 3))
    
    perf = pd.read_pickle(filename)[begin:end]
    fig = plt.figure(figsize=(15,10*(len(secs)+1)))
    ax1 = fig.add_subplot(len(secs)+1,1,1)
    perf.portfolio_value.plot(ax=ax1)
    perf.ending_cash.plot(ax=ax1)
    ax1.plot(data['SPY'].index, 1200*data['SPY']['price'])

    ax1.set_ylabel('portfolio value vs cash balance vs benchmark')

    # Display Indicators
    i = 1
    for sec in secs:
        i+=1;
        ax2 = fig.add_subplot(len(secs)+1,1,i)
        perf[sec].plot(ax=ax2)
        
        ax3 = ax2.twinx()
        perf[[sec+'-s_ma', sec+'-m_ma', sec+'-l_ma']].plot(ax=ax3)

        perf_trans = perf.ix[[t != [] for t in perf.transactions]]

        buys = perf_trans.ix[[(True in [(trade['sid']==sec and trade['amount'] > 0) for trade in t]) for t in perf_trans.transactions]]
        sells = perf_trans.ix[[(True in [(trade['sid']==sec and trade['amount'] < 0) for trade in t]) for t in perf_trans.transactions]]

        buy_markers = [max([(trade['amount'] if (trade['sid']==sec and trade['amount'] > 0) else 0) for trade in t]) for t in perf.transactions.ix[buys.index]]
        sell_markers = [-1*min([(trade['amount'] if (trade['sid']==sec and trade['amount'] < 0) else 0) for trade in t]) for t in perf.transactions.ix[sells.index]]

        ax2.scatter(buys.index, perf[sec].ix[buys.index], s=buy_markers, color='g', marker='^', alpha=.99)
        ax2.scatter(sells.index, perf[sec].ix[sells.index], s=sell_markers, color='r', marker='v', alpha=.99) 
        ax2.set_ylabel(sec+' price in $')

    plt.legend(loc=0)
    plt.show()
    
file_analyze(algo_name+'.pickle', 180)
```
