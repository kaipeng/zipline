```python
from zipline.algorithm import TradingAlgorithm
from zipline.api import order_target_value, record, symbol, history, add_history
import numpy as np
import pandas as pd

from collections import deque

algo_name = "Basic Momentum MVA Crossover - Assets - 3x-Asymmetric-Bloomberg-target_value"

# Define algorithm
def initialize(context):
    #set_benchmark(symbol('SPY'))
                  
    # Register 2 histories that track daily prices,
    # one with a 100 window and one with a 300 day window
    #add_history(10, '1d', 'price')
    add_history(50, '1d', 'price')
    add_history(100, '1d', 'price')
    add_history(200, '1d', 'price')

    context.i = 0
    context.secs = secs
    
    #context.window = pd.Panel(items=context.secs)
    #for sec in context.secs:
    context.cache = {}
    #context.cache['xs_ma'] = deque()
    context.cache['s_ma'] = deque()
    context.cache['m_ma'] = deque()
    context.cache['l_ma'] = deque()
        

def handle_data(context, data):
    # Skip first 300 days to get full windows
    context.i += 1
    if context.i < 200:
        return
    
    # Progess report
    if context.i % 100 == 0:
        print 'Bars Completed: ', context.i, data[data.keys()[0]]['dt'], 'Pos Value: ', context.portfolio.positions_value,'PnL: ', context.portfolio.pnl 

    # Compute averages
    # history() has to be called with the same params
    # from above and returns a pandas dataframe.
    short_mavg = history(50, '1d', 'price').mean()
    med_mavg = history(100, '1d', 'price').mean()
    long_mavg = history(200, '1d', 'price').mean()
    
    for sec in data.keys():
        
        ##### Trading logic ######
        # Compounding Moving Average Crossover Signals
        target = 0.0
        if short_mavg[sec] > med_mavg[sec]:
            target += .2
        if short_mavg[sec] > long_mavg[sec]:
            target += .3
        if med_mavg[sec] > long_mavg[sec]:
            target += .5

        if short_mavg[sec] < med_mavg[sec]:
            target -= .1
        if short_mavg[sec] < long_mavg[sec]:
            target -= .05
        if med_mavg[sec] < long_mavg[sec]:
            target -= 0
            
        target=target*(context.portfolio.portfolio_value)/len(data)
        ##### Trading logic ######
        
        # order_target orders as many shares as needed to
        # achieve the desired number of shares.
        order_target_value(sec, target)

        # Save values for later inspection
        record(sec, data[sec].price)
        record(sec+'-s_ma', short_mavg[sec])
        record(sec+'-m_ma', med_mavg[sec])
        record(sec+'-l_ma', long_mavg[sec])
    
    
    context.cache['s_ma'].append(short_mavg)
    context.cache['m_ma'].append(med_mavg)
    context.cache['l_ma'].append(long_mavg)
    
    #trim cache
    for key in context.cache.keys():
        if len(context.cache[key]) > 3:
            context.cache[key].popleft()
        

# Create algorithm object passing in initialize and
# handle_data functions
algo_obj = TradingAlgorithm(initialize=initialize, 
                            handle_data=handle_data)

# Run algorithm
data = pd.read_pickle(data_name+'.pickle')
perf_manual = algo_obj.run(data)
perf_manual.to_pickle(algo_name+'.pickle')
```

```python
def file_analyze(filename, begin=0, end=-1):
    data = load_bars_from_yahoo(stocks=['SPY'], start=startdate,
                            end=enddate)
    #data['SPY'].ix[:,:-2].plot(title='SPY', figsize=(15, 3))
    
    perf = pd.read_pickle(filename)[begin:end]
    fig = plt.figure(figsize=(15,10*(len(secs)+1)))
    ax1 = fig.add_subplot(len(secs)+1,1,1)
    perf.portfolio_value.plot(ax=ax1)
    perf.ending_cash.plot(ax=ax1)
    ax1.plot(data['SPY'].index, 1200*data['SPY']['price'])

    ax1.set_ylabel('portfolio value vs cash balance vs benchmark')

    # Display Indicators
    i = 1
    for sec in secs:
        i+=1;
        ax2 = fig.add_subplot(len(secs)+1,1,i)
        perf[sec].plot(ax=ax2)
        perf[[sec+'-s_ma', sec+'-m_ma', sec+'-l_ma']].plot(ax=ax2)

        perf_trans = perf.ix[[t != [] for t in perf.transactions]]

        buys = perf_trans.ix[[(True in [(trade['sid']==sec and trade['amount'] > 0) for trade in t]) for t in perf_trans.transactions]]
        sells = perf_trans.ix[[(True in [(trade['sid']==sec and trade['amount'] < 0) for trade in t]) for t in perf_trans.transactions]]

        buy_markers = [max([(trade['amount'] if (trade['sid']==sec and trade['amount'] > 0) else 0) for trade in t]) for t in perf.transactions.ix[buys.index]]
        sell_markers = [-1*min([(trade['amount'] if (trade['sid']==sec and trade['amount'] < 0) else 0) for trade in t]) for t in perf.transactions.ix[sells.index]]

        ax2.scatter(buys.index, perf[sec].ix[buys.index], s=buy_markers, color='g', marker='^', alpha=.99)
        ax2.scatter(sells.index, perf[sec].ix[sells.index], s=sell_markers, color='r', marker='v', alpha=.99) 
        ax2.set_ylabel(sec+' price in $')

    plt.legend(loc=0)
    plt.show()
algo_name = "Basic Momentum MVA Crossover - Assets - 3x-Asymmetric-Bloomberg-target_value"
file_analyze(algo_name+'.pickle', 2000,2050)
```
